```python
import pymysql
from tabulate import tabulate
import matplotlib.pyplot as plt

connection = pymysql.connect(
    user='root',
    password='root',
    host='localhost',
    database='library_app'
)
cursor = connection.cursor()


def add_member():
    print("Enter Member Details")
    try:
        member_id = int(input("Enter Member ID:"))
        Mname = input("Enter Member Name:")
        Madd = input("Enter Member Address:")
        phone = input("Enter Phone:")
        
        query = "INSERT INTO member VALUES (%s, %s, %s, %s)"
        values = (member_id, Mname, Madd, phone)
        cursor.execute(query, values)
        
        connection.commit()
        print("Insertion Success")
    except pymysql.connect.Error as error:
        print("Connection Lost:", error)

def edit_member():
    print("Edit Member details")
    try:
        member_id = int(input("Enter Member ID:"))
        query = "SELECT * FROM member WHERE member_id = %s"
        cursor.execute(query, (member_id,))
        pre_data = cursor.fetchone()
        
        if pre_data:
            member_name = input("Enter Member Name:")
            member_address = input("Enter Member Address:")
            phone = input("Enter Phone:")
            
            query = "UPDATE member SET Mname = %s, Madd = %s, phone = %s WHERE member_id = %s"
            values = (member_name, member_address, phone, member_id)
            cursor.execute(query, values)
            
            connection.commit()
            print("Successfully Updated")
        else:
            print("Member Not Found.")
    except ValueError:
        print("Invalid. Please check your Member ID.")
    except pymysql.connect.Error as error:
        print("Error in connection:", error)

def del_member():
    print("Delete Member Details")
    try:
        member_id = int(input("Enter the Member ID to delete:"))
        query = "SELECT * FROM member WHERE member_id = %s"
        cursor.execute(query, (member_id,))
        data = cursor.fetchone()
        
        if data:
            del_query = "DELETE FROM member WHERE member_id = %s"
            cursor.execute(del_query, (member_id,))
            
            connection.commit()
            print("Successfully Deleted")
        else:
            print("Member not found with the given ID.")
    except ValueError:
        print("Invalid. Enter a valid Member ID.")
    except pymysql.connect.Error as error:
        print("Error in connection:", error)

def search_member():
    print("Search Member by Member ID")
    try:
        member_id = int(input("Enter Member ID:"))
        query = "SELECT * FROM member WHERE member_id = %s"
        cursor.execute(query, (member_id,))
        result = cursor.fetchone()

        if result:
            headers = [i[0] for i in cursor.description]
            print(tabulate([result], headers=headers))
        else:
            print("Member not found with the given ID.")
    except ValueError:
        print("Invalid. Enter a valid Member ID.")
    except pymysql.connect.Error as error:
        print("Error in connection:", error)

        
def display_members():
    
    try:
        query = "SELECT * FROM member"
        cursor.execute(query)
        results = cursor.fetchall()
        headers = [i[0] for i in cursor.description]
        print(tabulate(results, headers=headers))
    except pymysql.connect.Error as error:
        print("Error in connection:", error)        
        

print("Done.")
```
