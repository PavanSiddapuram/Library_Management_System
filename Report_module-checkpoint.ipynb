```python
from tabulate import tabulate
import mysql.connector
import matplotlib.pyplot as plt

connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='root',
    database='library_app',
)

cursor = connection.cursor()

def display_book_details():
    print("Displaying Book Details:")
    try:
        query = "SELECT * FROM book"
        cursor.execute(query)
        results = cursor.fetchall()
        headers = [i[0] for i in cursor.description]
        print(tabulate(results, headers=headers, tablefmt="pretty"))
        print("Book Details retrieved successfully!")

    except mysql.connector.Error as error:
        print(f"MySQL error: {error}")

def display_member_details():
    print("Displaying Member Details:")
    try:
        query = "SELECT * FROM member"
        cursor.execute(query)
        results = cursor.fetchall()
        headers = [i[0] for i in cursor.description]
        print(tabulate(results, headers=headers, tablefmt="pretty"))
        print("Member Details retrieved successfully!")

    except mysql.connector.Error as error:
        print(f"MySQL error: {error}")

def display_book_issue_details():
    print("Displaying Book Issue Details:")
    try:
        query = "SELECT * FROM issue"
        cursor.execute(query)
        results = cursor.fetchall()
        headers = [i[0] for i in cursor.description]
        print(tabulate(results, headers=headers, tablefmt="pretty"))
        print("Book Issue Details retrieved successfully!")

    except mysql.connector.Error as error:
        print(f"MySQL error: {error}")

def display_book_return_details():
    print("Displaying Book Return Details:")
    try:
        query = "SELECT * FROM re_turn"
        cursor.execute(query)
        results = cursor.fetchall()
        headers = [i[0] for i in cursor.description]
        print(tabulate(results, headers=headers, tablefmt="pretty"))
        print("Book Return Details retrieved successfully!")

    except mysql.connector.Error as error:
        print(f"MySQL error: {error}")

def best_reading_books():
    print("Displaying Copies Issued for All Books:")
    try:
        # Query to get the copies issued for all books that have been issued
        query = "SELECT b.Bname, COALESCE(SUM(i.copies), 0) AS total_copies FROM book b INNER JOIN issue i ON b.book_id = i.book_id GROUP BY b.book_id, b.Bname ORDER BY total_copies DESC"
        cursor.execute(query)
        results = cursor.fetchall()

        if results:
            headers = ["Book Name", "Total Copies Issued"]
            print(tabulate(results, headers=headers, tablefmt="pretty"))

            # Creating vertical bar graph for all issued books
            titles = [result[0] for result in results]
            total_copies = [result[1] for result in results]

            plt.figure(figsize=(10, 8))
            plt.bar(titles, total_copies, color='green')
            plt.xlabel('Book Name')
            plt.ylabel('Total Copies Issued')
            plt.title('Copies Issued for All Books')
            plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
            plt.show()

            print("Vertical bar graph created!")

    except mysql.connector.Error as error:
        print(f"MySQL error: {error}")
        
print("Done.")
```
